```python
import pandas as pd
import numpy as np
import sqlite3
import itertools

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('../data-set/IDS_ALLCountries.xlsx')
len(df)
```

## Data Pre-Processing

```python
## Taking the columns out and storing it in a list
column_headers = list(df.columns.values)
print("The Column Header : ", column_headers)
```

```python
## Removing the spaces
for i in range(len(column_headers)):
    if isinstance(column_headers[i], str):
        column_headers[i] = column_headers[i].replace(' ', '_')  ## space to _
        column_headers[i] = column_headers[i].replace('-', '_') ## - to _
```

```python
## Spaces are removed
print("The Column Header :", column_headers)
```

```python
## Applying the modified columns back into the data-set
df.columns = column_headers
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
print("Number of Rows : ",df.shape[0])
print("Number of Columns : ",df.shape[1])
```

## 4) Totaling the amount of debt owed by the countries

```python
debt_years = []
for i in range(len(column_headers)):
    if isinstance(column_headers[i], int):
        debt_years.append(column_headers[i])

print(debt_years)
```

```python
query = 'select distinct Country_Code from all_countries1'
country_code = pd.read_sql(query, connection1)

country_code = country_code.values.tolist() ## data-frame to 2d list
country_code = list(itertools.chain.from_iterable(country_code)) ## 2d list to 1d list

print('Distinct Country Code : ')
print(country_code)
```

```python
## separating info according to the country
country_wise_info = []

for i in range(len(country_code)):
    query_with_country_code = "select * from all_countries1 where Country_Code like '%{}%'".format(country_code[i])
    ans = pd.read_sql(query_with_country_code, connection1)
    country_wise_info.append(ans)
```

```python
country_wise_info[1].head()
```

```python
## from the separated country taking only debt info
country_wise_debt_info = []

for i in range(len(country_code)):
    debt_years = country_wise_info[i].iloc[:,7:66] ## From 1970 to 2028
    country_wise_debt_info.append(debt_years)
```

```python
country_wise_debt_info[0] ## Afganistan country info
```

```python
country_wise_year_debt = []
country_wise_total_debt = []

for i in range(len(country_code)):
    year_debt = country_wise_debt_info[i].sum()
    
    country_wise_year_debt.append(year_debt)          ## country and their year wise debt
    country_wise_total_debt.append(sum(year_debt)) ## country and their total debt
```

```python
data_tuples = list(zip(country_code, distinct_countries, country_wise_total_debt))
total_debt   = pd.DataFrame(data_tuples, columns = ['Country_Code', 'Country_Name', 'Total_Debt'])
total_debt
```

```python
## Sorting the Total_Debt in descending order
total_debt = total_debt.sort_values(['Total_Debt'], ascending=[True])
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(18, 40))
 
# Horizontal Bar Plot
plt.barh(total_debt['Country_Name'], total_debt['Total_Debt'])
 
# setting label of y-axis
plt.ylabel("Countries")
 
# setting label of x-axis
plt.xlabel("Total Debt")

for index, value in enumerate(total_debt['Total_Debt']):
    plt.text(value, index, str(value))

plt.title("Countires --> Their Total Debt")
plt.show()
```

## 6) Average amount of debt across indicators

```python
## Taking out distinct debt
query = 'select distinct(Series_Code) from all_countries1'
distinct_debt_indicators = pd.read_sql(query, connection1)

distinct_debt_indicators
```

```python
series_code = distinct_debt_indicators.values.tolist() ## data-frame to 2d list
series_code = list(itertools.chain.from_iterable(series_code)) ## 2d list to 1d list
```

```python
print('Distinct Series Code : ',len(series_code))
```

```python
## separating info according to the country
country_wise_info = []
for i in range(len(country_code)):
    query_with_country_code = "select * from all_countries1 where Series_Code like '%{}%' ".format(series_code[i])
    ans = pd.read_sql(query_with_country_code, connection1)
    country_wise_info.append(ans)
```

```python
## from the separated country taking only debt info
country_wise_debt_info = []
for i in range(len(country_code)):
    debt_years = country_wise_info[i].iloc[:,7:66] ## From 1970 to 2028
    country_wise_debt_info.append(debt_years)
```

```python
series_wise_year_debt = []
series_wise_total_debt = []
for i in range(len(country_code)):
    year_debt = country_wise_debt_info[i].sum()
    
    series_wise_year_debt.append(year_debt)              ## Series Code and their year wise debt
    series_wise_total_debt.append((year_debt).mean()) ## Series Code and their average debt
```

```python
data_tuples = list(zip(series_code, series_wise_total_debt))
total_debt   = pd.DataFrame(data_tuples, columns=['Series_Code', 'Total_Debt'])
total_debt.head()
```

```python
## Sorting the Total_Debt in descending order
total_debt = total_debt.sort_values(['Total_Debt'], ascending=[True])
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(18, 40))
 
# Horizontal Bar Plot
plt.barh(total_debt['Series_Code'], total_debt['Total_Debt'])
 
# setting label of y-axis
plt.ylabel("Series Code")
 
# setting label of x-axis
plt.xlabel("Total Debt")

for index, value in enumerate(total_debt['Total_Debt']):
    plt.text(value, index, str(value))

plt.title("Series Code --> Their Total Debt")
plt.show()
```

```python
country_code  = list(high_principal_repayments['Country_Code'])
country_name = list(high_principal_repayments['Country_Name'])
total_amount   = list(high_principal_repayments.iloc[:,7:66].sum(axis=1))  ## row-wise addition
```

```python
data_tuples = list(zip(country_code, country_name, total_amount))
total_indicator_debt = pd.DataFrame(data_tuples, columns=['Country_Code', 'Country_Name', 'Total_Debt'])

total_indicator_debt
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(18, 40))
 
# Horizontal Bar Plot
plt.barh(total_indicator_debt['Country_Name'], total_debt['Total_Debt'])
 
# setting label of y-axis
plt.ylabel("Country Name")
 
# setting label of x-axis
plt.xlabel("Total Debt")

for index, value in enumerate(total_debt['Total_Debt']):
    plt.text(value, index, str(value))

plt.title("Highest amount of Principal Repayments")
plt.show()
```
